```
#Importando todas as bibliotecas que eu vou usar
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_linear_regression
import numpy as np
```

```
#Importando a base de dados que vou usar para treinar meu modelo
dados = 'train_cnprova.csv'
```

```
df = pd.read_csv(dados)
```

```
#Vou dar uma olhada na carinha da minha tabela
df.head()
```

```
#Eu li a documentação do exercício e percebi que, para treinar o modelo, as únicas variáveis que eram relavantes eram as notas dos alunos. Por isso, criei um novo df só com elas
dfn = df[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT']]
```

```
#A carinha do meu novo df
dfn.head()
```

```
#Eu não consigo treinar um modelo com valores nulos. Então, substituí os nulos por zeros, que seria a nota de quem teve a nota, de alguma forma, nula.
dfn.fillna(value=0, axis=0)
```

```
#Agora nenhuma das minhas entradas é nula
dfn.info()
```

```
#Decidir mudar o nome do meu df
notas = dfn
```

```
#Criei um data frame em que todos os nulos do notas estão preenchidos. Eu não conhecia o comando inplace = True na época, rs
df_notas = notas.fillna(0)
```

```
#Separando a base de treino e a de teste
X_treino, X_teste, y_treino, y_teste = train_test_split(df_notas.drop(columns='NU_NOTA_MT'), df_notas['NU_NOTA_MT'], test_size=0.3)
```

```
#Fazendo a regressão linear
lr = LinearRegression()
lr.fit(X_treino, y_treino) 
```

```
#Dando o predito:
predito = lr.predict(X_teste)
```

```
#Quero ver qual o valor do predito
print('R2', lr.score(X_teste, y_teste))
```

```
#Agora, vou importar o df de teste
dfteste = 'test_cnprova.csv'
```

```
#Vamos ler o df
teste = pd.read_csv(dfteste)
```

```
#Primeiro de tudo, eu vou criar um df igual o que foi treinado
df_teste = teste[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC']]
```

```
#Vou preencher os nulos dele
novo_teste = df_teste.fillna(0)
```

```
#Vou colocar minha coluna na tabela de teste
teste['NU_NOTA_MT'] = NU_NOTA_MT
```

```
#Minha tabela com os dados preditos está aqui. Eu acrescentei as colunas de número de inscrição e da presença nas provas de LC, que aconteceu no mesmo dia da de MT.
#Ou seja, quem faltou na prova de LC, faltou na de MT, portanto, não tem nota de MT também.
resposta = teste[['NU_INSCRICAO', 'TP_PRESENCA_LC', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT']]
```
